```python
import matplotlib.pyplot as plt

import numpy as np
import math

x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
sin_axes = fig.add_axes([0, 0, 1, 2])     # start at 0, 0 and length = 1 (height), 1 (width)
cos_axes = fig.add_axes([1, 0, 1, 1])     # start at 1, 1 and length = 1 (height), 1 (width)
sin_axes.plot(x, y1)
cos_axes.plot(x, y2)

plt.show()
```

**figsize**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(figsize = (10, 6))        # figsize (width, height)
sin_axes = fig.add_axes([0, 0, 1, 1])
cos_axes = fig.add_axes([1, 1, 1, 1]) 
sin_axes.plot(x, y1)
cos_axes.plot(x, y2)
plt.title("Test", fontdict=None, loc='center', pad=None)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#Task 1: 
#  Plot x, y1    and     x, y2
#  Generate the plot below

fig = plt.figure()
sin_axes = fig.add_axes([1, 0, 1, 2])    
cos_axes = fig.add_axes([0, 1, 1, 1])

sin_axes.plot(x, y1)
cos_axes.plot(x, y2)

plt.show()
```

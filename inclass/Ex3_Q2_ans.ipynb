```python
# Task 1

# import packages

# Requirement:
#       complete the following template
#       do not hard code the numbers



import pandas as pd
import numpy as np

stock_code = '000002' 
path_name = 'DATA'
csv_name = path_name + '/' +'stock-{}'.format(stock_code)  + '.csv'
```

```python
# Task 2

#read csv file and put it in a DataFrame

# Requirement:
#       2 lines (1 for print)

df = pd.read_csv(csv_name)
print(df.head())
```

**Data Exploration**

```python
# Task 3

#Show the size of the dataset

# Requirement:
#       1 line

print(f"{'stock-{}'.format(stock_code)  + '.csv'} has {df.shape[1]} features and {df.shape[0]} samples")
```

```python
# Task 4

#Show the statistics of the dataset

# Requirement:
#       4 lines
#       Do not hard code the numbers. Use functions to generate numbers

print("Close Price:")
print(f"Average = {round(df['close'].mean(), 2)}, Min = {round(df['close'].min(), 2)}, Max = {round(df['close'].max(), 2)}, SD = {round(df['close'].std(), 2)}" )
print("Volumn:")
print(f"Average = {round(df['volume'].mean(), 2)}, Min = {round(df['volume'].min(), 2)}, Max = {round(df['volume'].max(), 2)}, SD = {round(df['volume'].std(), 2)}" )
```

```python
# Task 5

#Store data (close column) in a Dictionary

# Requirement:
#       2 lines (one for print)
#       do not hard code the numbers

stock_summary = {
    'Closing price': {'Average':round(df['close'].mean(), 2), 
                      "Min":round(df['close'].min(), 2), 
                      "Max": round(df['close'].max(), 2), 
                      "SD": round(df['close'].std(), 2)}
}
print(stock_summary)
```

```python
# Task 6

#Check for missing values in all columns

# Requirement:
#       1 line
#       use isnull and shape
#       do not hard code the numbers

print(f'{df[df.isnull().any(axis=1)].shape[0]} missing data is found.')
```

```python
# Task 7

# Update the values in "volume" column 
#     e.g., update this number from 69.819518 to 69819.518

# Requirement:
#       2 lines (one line for print)

df['volume'] = df['volume'] * 1000
print(df.head())
```

**Save data**

```python
# Task 8

# Save the updated DataFrame to a new CSV file, named stock-000002_output.csv

# Requirement:
#       2 lines
#       filename = 'stock-000002_output.csv' (do not hard code. Use as many existing variable as possible to generate the name)
#       save the file to the current working directory
#       The next text cell shows the content of csv file (no need to show it in your program)
#       Note: Make sure the output format is exactly the same as the one being shown below


csv_output = 'stock-{}'.format(stock_code) + '_output'  + '.csv'
df.to_csv(csv_output, index=False)
```

```python
# Task 1

# read data from the CSV file, which is in the current working directory
# assign values to variables

# Requirement:
#       CSV file is placed in the current working directory
#       do not hard code the numbers and values
#       use as many existing variables as possible


import pandas as pd
import numpy as np

csv_file = "googleplaystore_v2.csv"
df = pd.read_csv(csv_file)                            # <- complete this one
print(f'Number of features = {df.shape[1]}, Number of samples = {df.shape[0]}')           # <- complete this one
print(df.head())                    # <- complete this one
```

```python
# Task 2

#Produce basic information about the dataset

# Requirement:
#       1 line


df.info()                                     # <---- one line of code
```

```python
# Task 3

#Count total number of missing values at each column

print("Count total number of missing values at each column")
print(f"{df.isnull().sum()}")                 # <- complete this one
```

```python
# Task 4

#Count total number of missing values in all columns

print(f"The total number of missing values in all columns = {df.isnull().sum().sum()}")              # <- complete this one
```

```python
# Task 5

# Store all categories (unique names) in a list


cat = list(df.Category.unique())              # <- complete this one
print(type(cat))                              # <- complete this one
print(f"The total number of categories = {len(cat)}")         # <- complete this one
print(f'The first one is {cat[0]}, and the last one is {cat[-1]}')        # <- complete this one
```

```python
# Task 6

# Which category of app is the MOST popular one (in term of the total number of apps in the category)?


most_cat = df.Category.value_counts()           # <- complete this one
print(most_cat)       
```

```python
# Task 7

# How many apps were in the category

print(f"There are {len(df.groupby(['Category']).groups['BUSINESS'])} apps in Business category.")          # <- complete this one
print(f"There are {len(df.groupby(['Category']).groups['EDUCATION'])} apps in Education category.")        # <- complete this one
```

```python
# Task 8

# Show the total numbers of Installs

print(df.groupby('Category')['Installs'].sum())             # <- complete this one
```

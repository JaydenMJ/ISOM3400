# Task 5

(a) For each platform, what is the maximum sales figures in NA, EU and JP? Use only one line to answer it.

(b) Apply `.plot(kind = 'bar')` to your answer in (a).

(c) For publisher `Nintendo`, show the maximum and total Global_Sales in each year. Use one line to answer this part as well.

(d) Apply `.plot(kind = 'line')` to your answer in (c).


```python
data.groupby('Platform')['NA_Sales'].max()
```

```python
#agg()
data.groupby('Platform').agg({'NA_Sales':max,'EU_Sales':max,'JP_Sales':max})
```

```python
data.groupby('Platform').agg({'NA_Sales':max,'EU_Sales':max,'JP_Sales':max}).plot(kind = 'bar')
```

```python
data[data['Publisher']=='Nintendo'].groupby('Year').agg({'Global_Sales':[max,sum]})
```

```python
data[data['Publisher']=='Nintendo'].groupby('Year').agg({'Global_Sales':[max,sum]}).plot(kind = 'line')
```
